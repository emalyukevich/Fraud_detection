```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from pathlib import Path

import sys
sys.path.append(str(Path().resolve().parent))

from src.utils import get_important_features_by_corr
```

```python
DATA_PATH = Path('../data/raw/creditcard.csv')
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
df['LogAmount'] = np.log1p(df['Amount'])
df.head()
```

```python
important_features = get_important_features_by_corr(df, threshold=0.09)
important_features += ['Hour', 'LogAmount']
```

```python
scaler = StandardScaler()
scaled_values = scaler.fit_transform(df[important_features])

df_scaled = pd.DataFrame(scaled_values, columns=important_features)
df_scaled['Class'] = df['Class'].values
```

```python
df_scaled.to_csv('../data/processed/df_scaled.csv', index=False)
```
